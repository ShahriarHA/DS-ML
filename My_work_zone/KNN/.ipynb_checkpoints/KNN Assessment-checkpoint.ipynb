# spliting dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Weight',axis=1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,train_size=0.70,random_state=42)
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
x_train['Gender_encoded'] = le.fit_transform(x_train.Gender)
```

```python
x_train.head()
```

```python
x_train.drop('Gender',axis=1,inplace=True)
```

```python
x_train.head()
```

```python
x_test['Gender_encoded'] = le.transform(x_test.Gender)
```

```python
x_test.head()
```

```python
x_test.drop('Gender',axis=1,inplace=True)
x_test.head()
```

# Least Squared Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.predict(x_test)
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
lr.predict([[8555.000000,0]])
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
lr_mse = mean_squared_error(y_test,y_pred)
lr_mse
```

```python
lr_model_ev = {
    'Training_score':lr.score(x_train,y_train)*100,
    'Test_score':lr.score(x_test,y_test)*100,
    'MSE_linearRegression': lr_mse
}
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x = lr_model_ev.keys(),y=lr_model_ev.values())
plt.show()
```

# KNN Regression

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn_lr = KNeighborsRegressor()
```

```python
knn_lr.fit(x_train,y_train)
```

```python
knn_train_score = knn_lr.score(x_train,y_train)*100
knn_train_score
```

```python
knn_test_score = knn_lr.score(x_test,y_test)*100
knn_test_score
```

```python
knn_mse = mean_squared_error(y_test,knn_lr.predict(x_test))
```

```python
knn_mse
```

```python
knn_lr_model_ev = {
    'Training_score':knn_train_score,
    'Test_score':knn_test_score,
    'MSE_linearRegression': knn_mse
}
```

```python
sns.barplot(x=knn_lr_model_ev.keys(),y=knn_lr_model_ev.values())
```

## Comparison of Linear Rgeression and KNN Regressor

```python
#plot
comparison_lr_knnLR = pd.DataFrame({
    'Models': ['Linear Regression','Linear Regression','Linear Regression','KNN Regression','KNN Regression','KNN Regression'],
    'Evaluation_components': ['Training score','Test score','MSE','Training score','Test score','MSE'],
    'Evaluation_components_value': [lr.score(x_train,y_train)*100,lr.score(x_test,y_test)*100,lr_mse,knn_train_score,knn_test_score,knn_mse]
    
})
```

```python
comparison_lr_knnLR
```

```python
sns.barplot(x=comparison_lr_knnLR.Evaluation_components,y=comparison_lr_knnLR.Evaluation_components_value,hue=comparison_lr_knnLR.Models)
```

```python
plt.figure(figsize=(15,7))
ax = sns.barplot(data=comparison_classifier_models,x='Evaluation components',y='Evaluation componets values',hue='Models')
for container in ax.containers:
    ax.bar_label(container, fmt='%.3f', padding=4,label_type='center')
```
