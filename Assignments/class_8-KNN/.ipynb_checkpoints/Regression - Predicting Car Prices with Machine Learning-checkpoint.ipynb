## Handling Missing Data using imputation method

```python
# Numeric columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
numeric_cols
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
categorical_cols
```

```python
numeric_missing = [col for col in numeric_cols if df[col].isnull().sum() > 0]
numeric_missing
```

```python
categorical_missing = [col for col in categorical_cols if df[col].isnull().sum() > 0]
categorical_missing
```

```python
# fill with mean numeric columns
df[numeric_missing[0]] = df[numeric_missing[0]].astype(float).fillna(df[numeric_missing[0]].astype(float).mean())
df.head()
```

```python
# fill with mode in categorical columns 
for col in categorical_missing:
    df[col] = df[col].fillna(df[col].mode()[0])
df.head()
```
